```python
import random
import matplotlib.pyplot as plt
```

```python
def random_numstring(l):
    string = []
    for i in range(l):
        value = random.randint(0,1)
        string.append(value)
    return string
```

```python
def closest_numstring(str1, str2):
    distance1 = sum(str1)
    distance2 = sum(str2)
    if distance1 >= distance2:
        #print("str1 won")
        return str1, distance1
    else:
        #print("str2 won")
        return str2, distance2
```

```python
def GA_num(l):
    iteration = 0
    fitness = []
    best_fit = 0
    
    string = random_numstring(l)
        
    probability = 1/l
    
    while(iteration != 1500 and best_fit != l):
        x = []
        for a in range(0, len(string)):
            random_number = random.random()
            if probability > random_number:
                if(string[a] == 1):
                    x.append(0)
                elif(string[a] == 0):
                    x.append(1)
            else:
                x.append(string[a])

        string, fit = closest_numstring(string, x)
        fitness.append(fit)
        if fit > best_fit:
            best_fit = fit
        iteration += 1
        
    print("Iteration: {}, Best_fit: {}, Last_fit: {}".format(iteration, best_fit, fitness[-1]))
    return fitness
    
    #print("Exceeded 100 iterations")
```

```python
def plot_runs_GA_num(n):
    for i in range(0, n):
        dists = GA_num(100)
        plt.plot(range(1, len(dists)+1), dists, label="run{}".format(i+1))

    plt.xlabel("Generation")
    plt.ylabel("Fitness score")
    plt.legend()
    plt.show()
```

```python
plot_runs_GA_num(1)
```

```python
plot_runs_GA_num(10)
```
